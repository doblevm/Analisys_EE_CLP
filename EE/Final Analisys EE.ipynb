```python
def Cp(any_list, lsl, usl):
    arr = any_list.ravel()
    sigma = any_list.std()
    return float(usl-lsl) / (6*sigma)
```

```python
def Cpk(any_list, lsl, usl):
    arr = any_list.ravel()
    sigma = any_list.std()
    m = any_list.mean()
    Cpu = float(usl-m) / (3*sigma)
    Cpl = float(m-lsl) / (3*sigma)
    Cpk = np.min([Cpu, Cpl])
    return Cpk
```

```python
print(Cp(df_Full["CLP"], 18000, 22000))
print(Cpk(df_Full["CLP"], 18000, 22000))
```

```python
print(Cp(df_Full["SERV"], 10000, 15000))
print(Cpk(df_Full["SERV"], 10000, 15000))
```

```python
print(Cp(df_Full["G.Aire"], 10000, 15000))
print(Cpk(df_Full["G.Aire"], 10000, 15000))
```
